```python
import re
import csv
import codecs
import os
import datetime
import fileinput


for fnamex in os.listdir('D:/ISB/Term1/PRAT2/Green_Cab_Full_Clean2/'):

    sfname = fnamex.replace('-out.csv','').replace('green_','')[0:4]
    f=os.path.join('D:/ISB/Term1/PRAT2/Green_Cab_Full_Clean2/', fnamex)
    fo=os.path.join('D:/ISB/Term1/PRAT2/Green_Cab_Full_Clean3/', sfname+'/'+fnamex)

    print(f)
    with fileinput.FileInput(f, inplace=False) as file, open(fo, 'w') as outfile:
        for line in file:
            x = re.split(r'\t', line)
            #print(x[0])
            #print('length: {}'.format(len(x)))
            if len(x) > 2:
                try:
                    #print(x[1])
                    #print(datetime.strptime(x[1],'%Y-%m-%d %H:%M:%s'))
                    x[1] = x[1].rstrip()
                    x[2] = x[2].rstrip()
                    pudow = datetime.datetime.strptime(x[1],'%Y-%m-%d %H:%M:%S').weekday()
                    dodow = datetime.datetime.strptime(x[2],'%Y-%m-%d %H:%M:%S').weekday()
                    puhour = datetime.datetime.strptime(x[1],'%Y-%m-%d %H:%M:%S').hour
                    dohour = datetime.datetime.strptime(x[2],'%Y-%m-%d %H:%M:%S').hour
                    
                    puhour = int(puhour)
                    dohour = int(dohour)
                    puhourt=99
                    dohourt=99
                    
                    if puhour >= 0 and puhour < 5:
                        puhourt = 1
                    elif puhour >=5 and puhour < 7:
                        puhourt = 2
                    elif puhour >=7 and puhour < 9:
                        puhourt = 3
                    elif puhour >=9 and puhour < 12:
                        puhourt = 4                          
                    elif puhour >=12 and puhour <= 14:
                        puhourt = 5
                    elif puhour >= 15 and puhour < 17:
                        puhourt = 6
                    elif puhour >= 17 and puhour < 19:
                        puhourt = 7      
                    elif puhour >= 19 and puhour < 21:
                        puhourt = 7                    
                    else:
                        puhourt = 8
                        
                    if dohour >= 0 and dohour < 5:
                        dohourt = 1
                    elif dohour >=5 and dohour < 7:
                        dohourt = 2
                    elif dohour >=7 and dohour < 9:
                        dohourt = 3
                    elif dohour >=9 and dohour < 12:
                        dohourt = 4                          
                    elif dohour >=12 and dohour <= 14:
                        dohourt = 5
                    elif dohour >= 15 and dohour < 17:
                        dohourt = 6
                    elif dohour >= 17 and dohour < 19:
                        dohourt = 7      
                    elif dohour >= 19 and dohour < 21:
                        dohourt = 7                    
                    else:
                        dohourt = 8
                    
                    
                    #print(pudow)
                    #print(dodow)
                    #print(puhourt)
                    #print(dohourt)
                    #print(line)
                    line = line.replace('\n','')
                    line = line + '\t'
                    line = line + str(pudow) + '\t' + str(dodow) + '\t' + str(puhourt) + '\t' + str(dohourt) + '\n'
                    #print('success?')
                    outfile.write(line)
                    line = ''

                    
                except:
                    print('an excep')
                    #print(line)
                    line = line.replace('\n','')
                    line = line + '\t' + 'PU DOW' + '\t' + 'Drop DOW' + '\t' + 'PU Timeslot' + '\t' + 'DO Timeslot' + '\n'
                    outfile.write(line)
                    line = ''
                    #print('an exception')

    
    file.close()
    outfile.close()
```
