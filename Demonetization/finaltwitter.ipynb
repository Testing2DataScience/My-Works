```python

import pandas as pd 
import os
import pickle
import datetime

df = pd.DataFrame(columns=['tweet_id','text','user_id','user_screen_name','user_name','created_at','retweets','favorites'])

for x in os.listdir('C:\\twitter_scrap-master\\saved_tweets\\All'):
    if x.endswith('.pkl'):
        f_name = os.path.join('C:\\twitter_scrap-master\\saved_tweets\\All',x)
        file = open(f_name, 'rb')
        data = pickle.load(file)
        file.close()
        datex =0
        for item in data: 
           #print (item)
            try:
               
                datex = int(int(str(item['created_at']).split('.')[0])/1000)
                df = df.append({
                                'tweet_id':item['tweet_id'],'text':item['text'],'user_id':item['user_id'],
                                'user_screen_name':item['user_screen_name'],'user_name':item['user_name'],
                                'created_at':datetime.datetime.fromtimestamp(datex),
                                'retweets':item['retweets'],
                                'favorites':item['favorites']
                               },ignore_index=True)
                
            except:
                print('There is an exception')

                #df.to_excel(writer,"tweetdump")
df.to_csv("C:\\twitter_scrap-master\\saved_tweets\\All\\dump.xlsx")
print (df)

```
