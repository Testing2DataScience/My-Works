## Custom downloader from AWS link 
* This is to avoid internet connection issues
* This code downloads from 2015 to 2018

```python
#### Green Cab Data (Full)
import csv
import urllib.request
import codecs

for hdnYear in range(2014,2018):
        for hdnMonth in range(1,13):
            url = 'https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_' + str(hdnYear) + '-' + str(hdnMonth).zfill(2) + '.csv'
            print(url)
            response = urllib.request.urlopen(url)

            cr = csv.reader(codecs.iterdecode(response, 'utf-8'))

            f=open('D:/ISB/Term1/PRAT2/Green_Cab_Full/green_'+ str(hdnYear) + '-' +str(hdnMonth).zfill(2) +'.csv','w',newline='')
            csvw = csv.writer(f, delimiter='\t',quotechar='|', quoting=csv.QUOTE_MINIMAL,dialect="excel", lineterminator="\n")
            i=0

            for row in cr:
                csvw.writerow(row)
                i = i + 1
                if i%500000 == 0:
                    print('Number of records processed is {}'.format(i))

            f.close()
            response.close()
```
