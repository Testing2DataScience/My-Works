### Identify the number of records in error

```python
import re
import csv
import codecs
import os
import datetime
import fileinput

x = os.listdir('D:/ISB/Term1/PRAT2/Green_Cab_Full_Clean5/')

for listd in x:
    #print(os.path.isdir(listd))
    if os.path.isdir(os.path.join('D:/ISB/Term1/PRAT2/Green_Cab_Full_Clean5/',listd)) == True:
        folname = os.path.join('D:/ISB/Term1/PRAT2/Green_Cab_Full_Clean5/',listd)
        #print(folname)
        for fnamex in os.listdir(folname):

            f=os.path.join(folname,fnamex)

            #print(f)
            distiszero = 0 
            twentysectravel = 0
            totrecs = 0
            with fileinput.FileInput(f, inplace=False) as file:
                for line in file:
                    totrecs = totrecs + 1
                    x = re.split(r'\t', line)

                    if int(x[21]) <= 20:
                        twentysectravel = twentysectravel + 1
                    else:
                        if float(x[7]) == 0:
                            distiszero = distiszero + 1





            if (distiszero > 0) or (twentysectravel > 0):
                print('for file name {}, of {} records, there are {} records with dist zero and {} records with less than 20 sec travel'.format(fnamex,totrecs,distiszero,twentysectravel))

            twentysectravel = 0
            distiszero = 0
            totrecs = 0

            file.close()
    #outfile.close()
```
